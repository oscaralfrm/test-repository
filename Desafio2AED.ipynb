```python
'''

Desafío de Programación de AED - Secuencia de Collatz o 3n + 1.

Partiendo de cualquier número positivo (discreto-natural) se realizan cálculos
necesarios para llegar al 1. 

Si el número es impar, se multiplica por 3 y se le suma 1.

Si el número es par, se vuelve a procesar y se divide a la mitad.

Se realizará de manera tal que al final de todo el ciclo condicional el valor
de n pase a ser 1, y se tienen las siguientes consideraciones:

- Los DISTINTOS VALORES que va tomando el número n mientras se va procesando
y transformando hasta llegar al 1, se conoce como la ÓRBITA DEL NÚMERO N.

- La CANTIDAD de valores que va tomando el número n mientras se procesa hasta
convertirse en el número 1, se conoce como la LONGITUD DE LA ÓRBITA DE N.

OBJETIVOS A CUMPLIR DEL DESAFÍO:

- SUBPROBLEMA #1: Cargar por teclado el número 'n'.  -- CHECK --
- SUBPROBLEMA #2:
Calcular y mostrar por pantalla todos los valores que toma la órbita de 'n'.
-- CHECK --
- SUBPROBLEMA #3:
Calcular y mostrar por pantalla el valor de la longitud de órbita de 'n'.
-- CHECK --
- SUBPROBLEMA #4:
Calcular y mostrar por pantalla el PROMEDIO de todos los valores de la órbita,
para esto hay que declarar una variable 'total_cant_orbita' que será la suma
total de todos los valores de la órbita, y dividirla por el número de longitud
de dicha órbita. El PROMEDIO NO debe ser truncado (mostrar resultado float).
-- CHECK --
- SUBPROBLEMA #5:
Calcular y mosrar por pantalla el MAYOR VALOR OBTENIDO EN LA ÓRBITA (mostrar
el número más grande hallado de dicha secuencia).
-- CHECK --

'''


def sector_inicio():
  # Se carga un número 'n' arbitrario que será procesado.
  # Subproblema 1
  n = int(input("Cargue un valor para n: "))
  return n


def determinar_mayor(a, b):
  # Calcular mayor
  if a > b:
    num_mayor = a
  else:
    num_mayor = b

  return num_mayor

def secuencia_de_collatz():
  n = sector_inicio()
  cant_orbita = [n]
  acum_valor_orbita = n
  long_orbita = 0
  mayor = n
  print(n)
  while n > 0 and n != 1:  # El bucle termina cuando llega a 1 el número.

    # Se propone ciclo while para realizar el procesamiento secuencial de n.
    # El número es par en esta instancia.

    if n % 2 == 0: 
      n = n / 2
      cant_orbita.append(n)
      acum_valor_orbita += n
      
    else:  # El número es impar en esta instancia.
      n = (3 * n) + 1 
      cant_orbita.append(n)
      acum_valor_orbita += n

    mayor = determinar_mayor(mayor, n)
    # Sector de Resultados

    # Subproblema 2

    print("El vector de los elementos pertenecientes a la órbita es: ",
          cant_orbita)
    
    # Subproblema 3

    total_cant_orbita = len(cant_orbita)
    print("La longitud total de elementos de la órbita es de: ", 
          int(total_cant_orbita))
    
    # Subproblema 4
    print(acum_valor_orbita / total_cant_orbita)

    # Subproblema 5
    print("El mayor valor de la orbita es: ",
          mayor)


    

secuencia_de_collatz()
```
